# 数据可视化入门

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import matplotlib
matplotlib.style.use('ggplot')
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('data/sample1.csv')
```

## 可视化模型

### 单个变量的特征

```python
np.log10(df['monthly_sales']).hist(bins=50)
```

### 两个变量的关系

```python
# plot
df.plot(y='comments',x='monthly_sales')
```

```python
# scatter
df.plot.scatter(x='monthly_sales',y='comments',logx=True, logy=True)
```
